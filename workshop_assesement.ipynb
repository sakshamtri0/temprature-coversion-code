```python
import tkinter as tk
from tkinter import filedialog
from PIL import Image, ImageTk
import os

class ImageViewerApp:
    def _init_(self, root):
        self.root = root
        self.root.title("Image Viewer")

        # Initialize variables
        self.current_image = None
        self.image_index = 0
        self.images = []

        # Create widgets
        self.file_label = tk.Label(root, text="No Image Selected")
        self.image_label = tk.Label(root)
        self.prev_button = tk.Button(root, text="Previous", state=tk.DISABLED, command=self.show_previous_image)
        self.next_button = tk.Button(root, text="Next", state=tk.DISABLED, command=self.show_next_image)
        self.browse_button = tk.Button(root, text="Browse", command=self.browse_images)

        # Grid layout
        self.file_label.grid(row=0, column=0, columnspan=2, padx=10, pady=10)
        self.image_label.grid(row=1, column=0, columnspan=2, padx=10, pady=10)
        self.prev_button.grid(row=2, column=0, padx=10, pady=10)
        self.next_button.grid(row=2, column=1, padx=10, pady=10)
        self.browse_button.grid(row=3, column=0, columnspan=2, padx=10, pady=10)

    def browse_images(self):
        filetypes = (("Image files", "*.jpg;*.jpeg;*.png;*.gif"), ("All files", "*.*"))
        image_files = filedialog.askopenfilenames(filetypes=filetypes)

        if image_files:
            self.images = list(image_files)
            self.image_index = 0
            self.load_image()
            self.update_navigation_buttons()

    def load_image(self):
        if self.images:
            image_path = self.images[self.image_index]
            self.current_image = Image.open(image_path)
            self.resize_image()
            self.update_image_label()
            self.update_file_label(os.path.basename(image_path))

    def resize_image(self):
        width, height = self.current_image.size
        max_width = self.root.winfo_width() - 20  # Subtracting padding
        max_height = self.root.winfo_height() - 60  # Subtracting padding and other widget heights

        # Scale the image to fit within the available space
        if width > max_width or height > max_height:
            ratio = min(max_width / width, max_height / height)
            new_width = int(width * ratio)
            new_height = int(height * ratio)
            self.current_image = self.current_image.resize((new_width, new_height), Image.ANTIALIAS)

    def update_image_label(self):
        image_tk = ImageTk.PhotoImage(self.current_image)
        self.image_label.configure(image=image_tk)
        self.image_label.image = image_tk

    def update_file_label(self, filename):
        self.file_label.configure(text=filename)

    def update_navigation_buttons(self):
        num_images = len(self.images)
        if num_images > 1:
            self.prev_button.configure(state=tk.NORMAL)
```

```python
import os
from tkinter import *
from PIL import Image, ImageTk

class ImageViewer:
    def __init__(self, master):
        self.master = master
        self.current_image_index = 0

        self.file_list = []
        self.load_file_list()

        self.image_label = Label(master)
        self.image_label.pack(pady=10)

        self.prev_button = Button(master, text="Previous", command=self.show_previous_image)
        self.prev_button.pack(side=LEFT, padx=10)

        self.next_button = Button(master, text="Next", command=self.show_next_image)
        self.next_button.pack(side=LEFT, padx=10)

        self.update_image()

    def load_file_list(self):
        directory = filedialog.askdirectory()
        if directory:
            self.file_list = [os.path.join(directory, file) for file in os.listdir(directory) if file.lower().endswith(('.jpg', '.jpeg', '.png'))]

    def update_image(self):
        if self.file_list:
            image_path = self.file_list[self.current_image_index]
            image = Image.open(image_path)
            image.thumbnail((400, 400))  # Adjust the size as needed
            photo = ImageTk.PhotoImage(image)
            self.image_label.config(image=photo)
            self.image_label.image = photo

    def show_previous_image(self):
        if self.current_image_index > 0:
            self.current_image_index -= 1
            self.update_image()

    def show_next_image(self):
        if self.current_image_index < len(self.file_list) - 1:
            self.current_image_index += 1
            self.update_image()

root = Tk()
root.title("Image Viewer")
app = ImageViewer(root)
root.mainloop()
```
